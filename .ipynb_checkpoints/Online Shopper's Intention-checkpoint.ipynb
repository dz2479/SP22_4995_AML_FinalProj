```python
import datetime

import pandas as pd
import numpy as np

import sklearn 

from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RandomizedSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import BernoulliNB 
from sklearn.svm import SVC
from sklearn.pipeline import Pipeline
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
from xgboost import XGBClassifier, plot_importance
#Metrics Libraries
from sklearn import metrics
from sklearn.metrics import classification_report
from sklearn.model_selection import cross_val_score

from scipy import stats
from scipy.stats import uniform, randint

#Configure Visualization Defaults
#%matplotlib inline = show plots in Jupyter Notebook browser
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib as mpl
import matplotlib.pylab as pylab
%matplotlib inline
mpl.style.use('ggplot')
pylab.rcParams['figure.figsize'] = 16,9

#beautiful progress bar
from tqdm import tqdm_notebook as tqdm

#ignore warnings
import warnings
warnings.filterwarnings('ignore')
```

It's easy to tell that **Informational**,  **informational_Duration** and **Administrative** contain almost none useful information, so we could drop them before establishing predictor.

On the other hand, **ProductRelated_Duration** and **ProductRelated**, **BounceRates** and **ExitRates** are highly related (reasonable), so we keep the **ProductRelated_Duration** and **ExitRates** only since the specific staying time preserve more details and exit rates is a better negative index.

```python
customer_copy.drop(["Informational","Informational_Duration","ProductRelated", "BounceRates"], axis = 1, inplace = True)
```

And then the outliers in **Administrative**, **Administrative_Duration**, **ProductRelated_Duration** should be removed by three-sigma method.

```python
customer_copy = customer_copy[(np.abs(stats.zscore(customer_copy.iloc[:, [0, 3]], axis=0)) < 3).all(axis=1)]
customer_copy.reset_index(inplace = True, drop = True)
```

```python
customer_copy.info()
```

Machine Learning models prefer numerical variables most of the time, thus we shall convert **Month**, **VisitorType**, **Weekend** and **Revenue** to int type.

```python
customer_copy.loc[:,"Month"].unique()
```

Oops! There is an unconventional abbreviation **"June"** that may cause trouble later, so we fix it first.

```python
def abbrtonum(month_str):
    if month_str == 'June':
        month_str = 'Jun'
    return datetime.datetime.strptime(month_str,"%b").month

customer_copy.loc[:,"Month"] = customer_copy.loc[:,"Month"].map(lambda x: abbrtonum(str(x)))
```

```python
customer_copy.loc[:,"VisitorType"].unique()
```

**VisitorType** has three different categorical values, so one-hot encoding method seems useful here.

```python
tmp_df = pd.get_dummies(customer_copy.VisitorType, prefix='VT')
customer_copy.drop(["VisitorType"], axis = 1, inplace = True)
customer_copy = pd.concat([customer_copy, tmp_df], axis=1)
```

```python
customer_copy.loc[:,"Weekend"] = customer_copy.loc[:,"Weekend"].map(lambda x: int(x))
customer_copy.loc[:,"Revenue"] = customer_copy.loc[:,"Revenue"].map(lambda x: int(x))
```

```python
customer_copy.info()
```

Now, It's time for Train and Test Spliting (70:30)

```python
x = customer_copy.drop(['Revenue'],axis=1)
y = customer_copy.Revenue

x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.7, random_state=4995, stratify=y)
print(y_train.value_counts())
print(y_test.value_counts())
```

By using Pipeline, we could test multiple models reproducibly and avoid some subtle misinformation. Based on our model plan, we may need several different pipelines. We will tune each model with grid search fashion and use accuracy as an criterion for model comparison.

Considering the buisness interests, we focus on those consumers who did bring profit to the website, so the F1 score is a better calibration for model selection - we will check that later.

```python
# for svc
step_svc = [("ss", StandardScaler()), ("pca", PCA()), ('svc', SVC())]
grid_para_svc = {"svc__C":[2,4,6,10],
                "svc__gamma":[0.01,0.025,0.05],
                "svc__kernel":['poly', 'rbf', 'sigmoid']}
pipline_svc = Pipeline(step_svc)
gridsearch_svc = GridSearchCV(pipline_svc, grid_para_svc, cv=4, scoring='accuracy', n_jobs=4)
gridsearch_svc.fit(x_train, y_train)
print("The accuracy score in cross validation is {0:.4f}".format(gridsearch_svc.best_score_))
print("The accuracy score in test data set is {0:.4f}".format(gridsearch_svc.score(x_test, y_test)))
print("The best parameters:\n {}".format(gridsearch_svc.best_params_))
```

```python
# for Random Forest
step_rf = [("ss", StandardScaler()), ('rf', RandomForestClassifier())]
grid_para_rf = {"rf__max_depth":[10,20,50],
                "rf__n_estimators":[10,50,100,200],
                "rf__min_samples_leaf":[1,2,5,10,15],
                "rf__max_leaf_nodes": [100,200,500,1000]}
pipline_rf = Pipeline(step_rf)
gridsearch_rf = GridSearchCV(pipline_rf, grid_para_rf, cv=4, scoring='accuracy')
gridsearch_rf.fit(x_train, y_train)
print("The accuracy score in cross validation is {0:.4f}".format(gridsearch_rf.best_score_))
print("The accuracy score in test data set is {0:.4f}".format(gridsearch_rf.score(x_test, y_test)))
print("The best parameters:\n {}".format(gridsearch_rf.best_params_))
```

```python
# for Navie Bayes
step_nb = [("ss", StandardScaler()), ('nb', BernoulliNB())]
grid_para_nb = {}
pipline_nb = Pipeline(step_nb)
gridsearch_nb = GridSearchCV(pipline_nb, grid_para_nb, cv=4, scoring='accuracy')
gridsearch_nb.fit(x_train, y_train)
print("The accuracy score in cross validation is {0:.4f}".format(gridsearch_nb.best_score_))
print("The accuracy score in test data set is {0:.4f}".format(gridsearch_nb.score(x_test, y_test)))
print("The best parameters:\n {}".format(gridsearch_nb.best_params_))
```

```python
# for logistic regression
step_lr = [("ss", StandardScaler()), ("pca", PCA()), ('lr', LogisticRegression(random_state=4995))]
grid_para_lr = {"lr__penalty":['l1','l2','elasticnet','none'],
               "lr__C":[0.5,1,2],
               "lr__solver":['newton-cg','lbfgs','liblinear','sag', 'saga'],
               "lr__intercept_scaling":[1,2,3,4]}
pipline_lr = Pipeline(step_lr)
gridsearch_lr = GridSearchCV(pipline_lr, grid_para_lr, cv=4, scoring='accuracy')
gridsearch_lr.fit(x_train, y_train)
print("The accuracy score in cross validation is {0:.4f}".format(gridsearch_lr.best_score_))
print("The accuracy score in test data set is {0:.4f}".format(gridsearch_lr.score(x_test, y_test)))
print("The best parameters:\n {}".format(gridsearch_lr.best_params_))
```

```python
# for Xgboost
step_xgb = [("ss", StandardScaler()), ('xgb', XGBClassifier(random_state=4995))]
grid_para_xgb = {    
                "xgb__colsample_bytree": uniform(0.7, 0.3),
                "xgb__gamma": uniform(0, 0.5),
                "xgb__learning_rate": uniform(0.03, 0.3),
                "xgb__max_depth": randint(2, 6),
                "xgb__n_estimators": randint(100, 150),
                "xgb__subsample": uniform(0.6, 0.4)
                }
pipline_xgb = Pipeline(step_xgb)
# We use randomized search method here, but still name it as 'grid' for homogeneity
gridsearch_xgb = RandomizedSearchCV(pipline_xgb, grid_para_xgb, cv=4, scoring='accuracy', n_jobs=4, random_state=4995, n_iter=200,)
gridsearch_xgb.fit(x_train, y_train)
print("The accuracy score in cross validation is {0:.4f}".format(gridsearch_xgb.best_score_))
print("The accuracy score in test data set is {0:.4f}".format(gridsearch_xgb.score(x_test, y_test)))
print("The best parameters:\n {}".format(gridsearch_xgb.best_params_))
```

Noting that the best accuracy in test data set is 0.8992 in wake of Xgboost Model, it seems to be our final choice. 

Let's Check the feature importance based on Xgboost model.

```python
feature_importance = gridsearch_xgb.best_estimator_.named_steps['xgb'].feature_importances_
feat_importances = pd.Series(feature_importance, index = x_train.columns)
feat_importances.nlargest(15).plot(kind = 'bar', colormap = 'PuOr')
plt.show()
```

```python
y_pred = gridsearch_xgb.best_estimator_.named_steps['xgb'].predict(x_test)
print("Classification Report:\n",classification_report(y_test,y_pred))
```

The f1-score of positive observation is 0.42, which is OK. And the precision is 0.79, which is fabulous! However, according to our practical experience, these models are too good to be true with incredible high precision. We are afraid there is some thing misleading.

After reinspect the independants, we take notice of **PageValues** which is intuitively different from other variables. It is highly possible that **PageValues** is a derivative feature rather than a primitive one, and the correlation heatmap do tell us that **PageValues** evidently related with **Revenue**. We guess that **PageValues** was derived from **Revenue** for data analyze, just like many production teams ordinary practices, representing high convert rate. Thus, to avoid inverting cause and effect, getting rid of this variable and redo our models is worthwhile.

```python
x_train.drop("PageValues", axis = 1, inplace = True)
x_test.drop("PageValues", axis = 1, inplace = True)
```

```python
# reuse what coded
model_list = [gridsearch_svc, gridsearch_rf, gridsearch_nb, gridsearch_lr, gridsearch_xgb]
model_names = {0:'gridsearch_svc', 1:'gridsearch_rf', 2:'gridsearch_nb', 3:'gridsearch_lr', 4:'gridsearch_xgb'}
for idx, model in tqdm(enumerate(model_list)):
    model.fit(x_train, y_train)
    print(f"---- {model_names[idx]} ----")
    print("The accuracy score in cross validation is {0:.4f}".format(model.best_score_))
    print("The accuracy score in test data set is {0:.4f}".format(model.score(x_test, y_test)))
    print("The best parameters:\n {}".format(model.best_params_))
```

So every model except Navie Bayes has close accuracy score. Let's check the feature importance and confusion matrix of Xgboos as well.

```python
feature_importance = gridsearch_xgb.best_estimator_.named_steps['xgb'].feature_importances_
feat_importances = pd.Series(feature_importance, index = x_train.columns)
feat_importances.nlargest(15).plot(kind = 'bar', colormap = 'PuOr')
plt.show()
```

```python
y_pred = gridsearch_xgb.best_estimator_.named_steps['xgb'].predict(x_test)
print("Classification Report:\n",classification_report(y_test,y_pred))
```

Well, after removing **PageValue**, the *precision* and *recall* of positive observation in test data set is ZERO, which means we are totally incapable to predict customers behavior with given variables. It's unsettled but understandable.
